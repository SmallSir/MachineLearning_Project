```python
import lightgbm as lgb
from sklearn.pipeline import Pipeline
from sklearn import metrics
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
import time
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import GridSearchCV
from imblearn.combine import SMOTETomek
```

```python
tests = pd.read_table('/content/gdrive/My Drive/keda/test.txt')
trains = pd.read_table('/content/gdrive/My Drive/keda/train.txt')
```

```python
trains['advert_industry_first'] = trains.loc[:,'advert_industry_inner'].str.split('_',expand = True)[0]
trains['advert_industry_second'] = trains.loc[:,'advert_industry_inner'].str.split('_',expand = True)[1]
tests['advert_industry_first'] = tests.loc[:,'advert_industry_inner'].str.split('_',expand = True)[0]
tests['advert_industry_second'] = tests.loc[:,'advert_industry_inner'].str.split('_',expand=True)[1]
```

```python
dele_feature = ['inner_slot_id','f_channel','advert_industry_inner','app_paid','creative_is_voicead','creative_is_js','user_tags','os_name','make','model','advert_name','osv','time']
train = trains.drop(dele_feature,axis = 1)
test = tests.drop(dele_feature,axis = 1)
```

```python
data = pd.concat([train,test],axis = 0)
data = data.reset_index(drop = True)
```

```python
def solve_appid(da):
    if da['app_id'] == 0:
         return data[data['advert_id'] == da['advert_id']]['app_id'].mode()
    else:
        return da['app_id']
train['app_id'].fillna(0,inplace =True)
train['app_id'] = train[['advert_id','app_id']].apply(solve_appid,axis = 1)
test['app_id'].fillna(0,inplace =True)
test['app_id'] = test[['advert_id','app_id']].apply(solve_appid,axis = 1)
```

```python
def solve_appcateid(da):
    if da['app_cate_id'] == 0:
        return data[data['advert_id'] == da['advert_id']]['app_cate_id'].mode()
    else:
        return da['app_cate_id']
train['app_cate_id'].fillna(0,inplace = True)
train['app_cate_id'] = train[['app_cate_id','advert_id']].apply(solve_appcateid,axis = 1)
test['app_cate_id'].fillna(0,inplace = True)
test['app_cate_id'] = test[['app_cate_id','advert_id']].apply(solve_appcateid,axis = 1)
```

```python
data = pd.concat([train,test],axis = 0)
data = data.reset_index(drop = True)
```

```python
data['creative_area'] = data['creative_height'] * data['creative_width']
```

```python
train['creative_area'] = train['creative_height'] * train['creative_width']
test['creative_area'] = test['creative_height'] * test['creative_width']
```

```python
corr = data.corr()
plt.subplots(figsize=(18, 16)) # 设置画面大小
sns.heatmap(corr, annot=True, vmax=1, square=True, cmap="Blues")
plt.show()
```

```python
onehot = ['advert_industry_second','advert_industry_first','province','os','nnt','devtype','creative_width','creative_type','creative_tp_dnf','creative_height','city','carrier','campaign_id','creative_area']
```

```python
for col in onehot:
    data[col] = data[col].astype('category')
```

```python
tt = pd.get_dummies(data[onehot])
```

```python
int_name = ['adid','advert_id','creative_id','orderid']
object_name = ['app_cate_id','app_id']
float_name = ['click']
for col in int_name:
    data[col] = data[col].astype('int32')
for col in object_name:
    data[col] = data[col].astype('float')
```

```python
da = pd.concat([tt,data],axis = 1)
```

```python
da.drop(onehot,inplace=True,axis = 1)
```

```python
data = da
train = data.iloc[:1001650,:]
test = data.iloc[1001650:,:]
```

```python
#调参参数
pargm_grid = {
    'learning_rate': [0.02,0.04,0.06,0.08,0.1],
    'max_depth': [-1,2,4,6,8]
}
drop_list = ['instance_id','click']
#设置训练集
train_x = train.drop(drop_list,axis = 1)
train_y = train['click']
#设置测试集
test_x = test.drop(drop_list,axis = 1)
```

```python
lg = lgb.LGBMClassifier()
searchCV = GridSearchCV(lg,pargm_grid,cv = 10,scoring="neg_log_loss")
searchCV.fit(train_x,train_y)
```

```python
searchCV.best_params_
```

```python
train_xx,Train_x,train_yy,Train_y = train_test_split(train_x,train_y,test_size = 0.1)

clf = lgb.LGBMClassifier(
        boosting_type='gbdt', num_leaves=64, reg_alpha=0.0, reg_lambda=1,
        max_depth=-1, n_estimators=3000, objective='binary',
        subsample=0.7, colsample_bytree=0.7, subsample_freq=1,
        learning_rate=0.1, min_child_weight=50,random_state=2018,n_jobs=-1
    )

clf.fit(train_x, train_y,eval_set=[(train_xx, train_yy),(Train_x,Train_y)],early_stopping_rounds=100,verbose=10)
```

```python
prob_submit = clf.predict_proba(test_x,num_iteration=clf.best_iteration_)[:,1]
sub = pd.DataFrame({'instance_id':test['instance_id'].values.tolist(),'predicted_score':prob_submit.tolist()})
sub.to_csv('baseline1.csv',index=False)
```
